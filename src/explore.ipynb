```python
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
```

```python
cat_dir = r'C:\Users\wipip\OneDrive\Documentos\GitHub\Clasificador-de-imagenes--main\data\organized\cat'
dog_dir = r'C:\Users\wipip\OneDrive\Documentos\GitHub\Clasificador-de-imagenes--main\data\organized\dog'
```

```python
cat_images = [os.path.join(cat_dir, img) for img in os.listdir(cat_dir)[:9]]
dog_images = [os.path.join(dog_dir, img) for img in os.listdir(dog_dir)[:9]]
```

```python
def plot_images(images, title):
    plt.figure(figsize=(10, 10))
    for i, img_path in enumerate(images):
        img = image.load_img(img_path, target_size=(200, 200))  # Ajustar el tama√±o si es necesario
        plt.subplot(3, 3, i + 1)
        plt.imshow(img)
        plt.title(title)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plot_images(cat_images, "Gato")
```

```python
plot_images(dog_images, "Perro")
```
